```python
# 샘플 데이터

# 문제

# 1) 과일별 판매량을 막대그래프로 그려 보세요.
#   - x축: 과일
#   - y축: 판매량

# 2) 과일별 가격 변화를 선그래프로 그려 보세요.
#   - x축: 과일
#   - y축: 가격

# 3) 판매량과 가격을 함께 고려한 "총매출(판매량*가격)" 열을 추가하고,
#    총매출을 나타내는 파이 차트를 그려 보세요.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'

# 샘플 데이터
data = {
    "과일": ["사과", "바나나", "포도", "딸기", "오렌지"],
    "판매량": [30, 45, 10, 25, 20],
    "가격": [1000, 500, 2000, 1500, 1200]
}
df = pd.DataFrame(data)

# 문제

# 1) 과일별 판매량을 막대그래프로 그려 보세요.
#   - x축: 과일
#   - y축: 판매량
fruit_list = list(df['과일'])
sales_list = list(df['판매량'])
price_list = list(df['가격'])
fig, ax = plt.subplots(2,2, facecolor='white', figsize=(10,10))

ax[0,0].bar(fruit_list, sales_list)
ax[0,0].set_xlabel('과일')
ax[0,0].set_ylabel('판매량', rotation = 'horizontal')

# 2) 과일별 가격 변화를 선그래프로 그려 보세요.
#   - x축: 과일
#   - y축: 가격

ax[0,1].plot(fruit_list, price_list)
ax[0,1].set_xlabel('과일')
ax[0,1].set_ylabel('가격', rotation = 'horizontal')

# 3) 판매량과 가격을 함께 고려한 "총매출(판매량*가격)" 열을 추가하고,
#    총매출을 나타내는 파이 차트를 그려 보세요.
total_revenues_list = []
for i in range(len(sales_list)):
    total_revenues_list.append(sales_list[i]*price_list[i])
ax[1,0].pie(total_revenues_list, labels = fruit_list)
```

```python
#조금 더 정리된 버전
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

data = {
    "과일": ["사과", "바나나", "포도", "딸기", "오렌지"],
    "판매량": [30, 45, 10, 25, 20],
    "가격": [1000, 500, 2000, 1500, 1200]
}
df = pd.DataFrame(data)
df["총매출"] = df["판매량"] * df["가격"]

fig, ax = plt.subplots(2, 2, figsize=(10, 10), facecolor='white')

# 1) 막대그래프
ax[0,0].bar(df["과일"], df["판매량"])
ax[0,0].set_xlabel("과일")
ax[0,0].set_ylabel("판매량")

# 2) 선그래프
ax[0,1].plot(df["과일"], df["가격"], marker="o")
ax[0,1].set_xlabel("과일")
ax[0,1].set_ylabel("가격")

# 3) 파이차트
ax[1,0].pie(df["총매출"], labels=df["과일"], autopct="%1.1f%%")
ax[1,0].set_title("총매출 비중")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

# 샘플 데이터 (학생 시험 점수)
data = {
    "이름": ["철수", "영희", "민수", "지민", "수현"],
    "국어": [82, 90, 76, 88, 95],
    "수학": [91, 85, 79, 92, 87],
    "영어": [78, 93, 84, 80, 89]
}
df = pd.DataFrame(data)

# 문제
# 1) 학생별 세 과목 평균 점수를 계산해서 새로운 열 '평균'을 추가하세요.
df["평균"] = df[["국어", "수학", "영어"]].mean(axis=1)

# 2) 학생별 평균 점수를 막대그래프로 그리세요.
fig, ax = plt.subplots(2,2, facecolor='white', figsize=(10,10))
ax[0,0].bar(df['이름'],df['평균'])
ax[0,0].set_xlabel('이름')
ax[0,0].set_ylabel('평균', rotation = 'horizontal')

# 3) 과목별 점수 분포(국어, 수학, 영어)를 한 그래프에 선그래프로 겹쳐서 그리세요.
#    (x축: 학생 이름, y축: 점수)
ax[0,0].plot(df['이름'],df['국어'], label = '국어', marker = 'o')
ax[0,0].plot(df['이름'],df['수학'], label = '수학', marker = '^')
ax[0,0].plot(df['이름'],df['영어'], label = '영어', marker = 's')
ax[0,0].legend()

# 4) 평균 점수를 기준으로 파이차트를 만들어 학생별 비중을 시각화하세요.
ax[0,1].pie(df['평균'], labels = df['이름'], autopct="%1.1f%%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

data = {
    "이름": ["철수", "영희", "민수", "지민", "수현"],
    "국어": [82, 90, 76, 88, 95],
    "수학": [91, 85, 79, 92, 87],
    "영어": [78, 93, 84, 80, 89]
}
df = pd.DataFrame(data)
df["평균"] = df[["국어", "수학", "영어"]].mean(axis=1)

fig, ax = plt.subplots(1, 2, figsize=(12, 5), facecolor="white")

# (왼쪽) 평균 막대 + 과목별 선그래프를 같은 축에 겹쳐서
ax[0].bar(df["이름"], df["평균"], color="lightgray", label="평균")
ax[0].plot(df["이름"], df["국어"], marker="o", label="국어")
ax[0].plot(df["이름"], df["수학"], marker="^", label="수학")
ax[0].plot(df["이름"], df["영어"], marker="s", label="영어")
ax[0].set_title("학생별 평균과 과목 점수")
ax[0].set_xlabel("이름")
ax[0].set_ylabel("점수")
ax[0].legend()

# (오른쪽) 평균 파이차트
ax[1].pie(df["평균"], labels=df["이름"], autopct="%1.1f%%", startangle=90)
ax[1].set_title("평균 점수 비중")

plt.tight_layout()
plt.show()
```

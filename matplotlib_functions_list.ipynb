```python
#figure: 전체 그래프 영역
#Axes: 그래프 생성 영역
#하나의 figure에 여러개의 Axes를 생성할 수 있음
#x축을 x axis, y축을 y axis라고 부름
#각각의 축의 라벨은 X axis label, Y axis label이라고 부름
#X축의 눈금을 X tick, Y축의 눈금을 Y tick이라고 부름
#그래프 제목은 title, 그래프 설명은 legend라고 부름
#그래프 위에 눈금은 grid라고 부름
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.rcParams['font.size'] =15 #글자 크기, 기본값은 10


fig, ax = plt.subplots(facecolor='white')
ax.plot(['월','화','수','목','금'],[10,60,30,20,50],
        label='고가', marker='o') #라벨 표시, 마커 표시
ax.plot(['월','화','수','목','금'],[5,40,25,10,30],
        label='저가', marker='^') #라벨 표시, 마커 표시, 사각형으로 하고싶을 경우에는 's'
#ax.set_xlabel('요일')
#ax.set_ylabel('가격', rotation='horizontal') #y축 라벨(판매량 글자자) 회전, 수직으로 하고 싶을 경우 rotation='vertical'
#ax.set_yticks([0,10,20,30,40,50,60,70]) #y축 눈금 설정
ax.set_title('주간 주식 변동성', fontsize=20) #그래프 제목, 글자 크기 20
ax.set(title='주간 주식 변동성', xlabel='요일', ylabel='가격', yticks=[0,10,20,30,40,50,60,70]) # 위 주석처리 3개를 이걸로 대체할 수 있음
#그러나 폰트사이즈 개별로 바꾸려면 위에있는 것처럼
ax.grid(True) #그래프 위에 눈금 표시
ax.legend() #라벨 표시
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.rcParams['font.size'] =15


fig, ax = plt.subplots(facecolor='white')
ax.bar(['월','화','수','목','금'],[10,60,30,20,50],
        label='고가') #봉으로 변경
ax.plot(['월','화','수','목','금'],[5,40,25,10,30],
        label='저가', marker='^', color='m') #색 변경, m은 마젠다
#ax.set_xlabel('요일')
#ax.set_ylabel('가격', rotation='horizontal') 
#ax.set_yticks([0,10,20,30,40,50,60,70]) 
ax.set(xlabel='요일', ylabel='가격', yticks=[0,10,20,30,40,50,60,70]) # 주석처리한 부분 간편화, 세로부분을 가로로 바꾸려면 위에 주석처리부분 써야함
ax.set_title('주간 주식 변동성', fontsize=20)
ax.grid(True)
ax.legend()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

fig, ax = plt.subplots(2,2,facecolor = 'white', figsize=(15,10)) #지정된 (행, 열) 안에 각각의 그래프가 들어감, figsize는 배경(지금은 흰색배경) 크기기

ax[0,0].bar(['월','화','수','목','금'],[10,60,30,20,50],
        label='고가')
ax[0,0].plot(['월','화','수','목','금'],[5,40,25,10,30],
        label='저가', marker='^', color='m')
ax[0,0].set_title('axes[0,0] 그래프')

ax[0,1].scatter([23,44,30,55,20,23],
           [12,34,23,22,13,33])
ax[0,1].set_title('axes[0,1] 그래프')

ax[1,0].hist([32,63,65,37,87,45,67,54,23,53,54,65,18,53,61,95,14,86,15,28,73],
        bins=11)
ax[1,0].set_title('axes[1,0] 그래프')

fig.savefig('graph.png') #이걸로 파일과 같은 경로에 'graph.png'로 파일 생성
```

```python
#삼각함수 그래프 그리기기
import numpy as np
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)

fig, ax = plt.subplots()
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]
ax.plot(X_1, Y_1)
ax.set(title='cos graph', xlabel='X', ylabel='Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


X = range(100)
Y = range(100)
X_1 = range(100)
Y_1 = [np.cos(value) for value in X] #코사인 그래프
X_2 = range(100)
Y_2 = [np.sin(value) for value in X] #사인 그래프


fig = plt.figure()
fig.set_size_inches(10, 10) #10, 10사이즈의 배경을 생성
ax_1 = fig.add_subplot(1, 2, 1) #이건 왼쪽에, 행이1 / 열이 2인 그래프프
ax_2 = fig.add_subplot(1, 2, 2) #이건 오른쪽에
ax_1.plot(X_1, Y_1, c='b') #파란색
ax_2.plot(X_2, Y_2, c='g') #초록색
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)
ax[0,0].plot(x, y_1)
ax[0,1].plot(x, y_2)
ax[1,0].plot(x, y_3)
ax[1,1].plot(x, y_4)
plt.show()
```

```python
#subplot 행렬
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

ax1 = plt.subplot(321)#321은 3,2,1. 즉, 3행, 2열로 나눴을 때 1번째 공간에 넣겠다
plt.plot(x, y_1)

ax2 = plt.subplot(322)#3행, 2열로 나눴을 때 2번째 공간에 넣겠다
plt.plot(x, y_1)

ax3 = plt.subplot(312)#3행, 1열로 나눴을 때 2번째 공간에 넣겠다
plt.plot(x, y_1)

ax4 = plt.subplot(325)#3행, 2열로 나눴을 때 5번째 공간에 넣겠다
plt.plot(x, y_1)

ax5 = plt.subplot(326)#3행, 2열로 나눴을 때 6번째 공간에 넣겠다
plt.plot(x, y_1)

plt.show()
```

```python
#figure: 전체 그래프 영역
#Axes: 그래프 생성 영역
#하나의 figure에 여러개의 Axes를 생성할 수 있음
#x축을 x axis, y축을 y axis라고 부름
#각각의 축의 라벨은 X axis label, Y axis label이라고 부름
#X축의 눈금을 X tick, Y축의 눈금을 Y tick이라고 부름
#그래프 제목은 title, 그래프 설명은 legend라고 부름
#그래프 위에 눈금은 grid라고 부름
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.rcParams['font.size'] =15 #글자 크기, 기본값은 10


fig, ax = plt.subplots(facecolor='white')
ax.plot(['월','화','수','목','금'],[10,60,30,20,50],
        label='고가', marker='o') #라벨 표시, 마커 표시
ax.plot(['월','화','수','목','금'],[5,40,25,10,30],
        label='저가', marker='^') #라벨 표시, 마커 표시, 사각형으로 하고싶을 경우에는 's'
ax.set_xlabel('요일')
ax.set_ylabel('가격', rotation='horizontal') #y축 라벨(판매량 글자자) 회전, 수직으로 하고 싶을 경우 rotation='vertical'
ax.set_yticks([0,10,20,30,40,50,60,70]) #y축 눈금 설정
ax.set_title('주간 주식 변동성', fontsize=20) #그래프 제목, 글자 크기 20
ax.grid(True) #그래프 위에 눈금 표시
ax.legend() #라벨 표시
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.rcParams['font.size'] =15


fig, ax = plt.subplots(facecolor='white')
ax.bar(['월','화','수','목','금'],[10,60,30,20,50],
        label='고가') #봉으로 변경
ax.plot(['월','화','수','목','금'],[5,40,25,10,30],
        label='저가', marker='^', color='m') #색 변경, m은 마젠다
ax.set_xlabel('요일')
ax.set_ylabel('가격', rotation='horizontal') 
ax.set_yticks([0,10,20,30,40,50,60,70]) 
ax.set_title('주간 주식 변동성', fontsize=20)
ax.grid(True)
ax.legend()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

fig, ax = plt.subplots(2,2,facecolor = 'white', figsize=(15,10)) #지정된 (행, 열) 안에 각각의 그래프가 들어감, figsize는 배경(지금은 흰색배경) 크기기

ax[0,0].bar(['월','화','수','목','금'],[10,60,30,20,50],
        label='고가')
ax[0,0].plot(['월','화','수','목','금'],[5,40,25,10,30],
        label='저가', marker='^', color='m')
ax[0,0].set_title('axes[0,0] 그래프')

ax[0,1].scatter([23,44,30,55,20,23],
           [12,34,23,22,13,33])
ax[0,1].set_title('axes[0,1] 그래프')

ax[1,0].hist([32,63,65,37,87,45,67,54,23,53,54,65,18,53,61,95,14,86,15,28,73],
        bins=11)
ax[1,0].set_title('axes[1,0] 그래프')

fig.savefig('graph.png') #이걸로 파일과 같은 경로에 'graph.png'로 파일 생성
```

```python
#불 인덱스, 특정조건하에서 추출
import pandas as pd

df = pd.read_csv('pandas_data_csv.csv')

#and조건
over_10b = df[(df['시가총액']>=10000000000) & (df['시장구분']=='KOSPI')& (df['거래량']>=10000)]
#print(over_10b)

#or조건
DAQPI = df[(df['시장구분']=='KOSPI') | (df['시장구분']=='KOSDAQ')]
print(DAQPI)
```

```python
#통계, 연산
import pandas as pd

df = pd.read_csv('pandas_data_csv.csv')
df['시가총액'].mean() #평균, numeric_only=True는 
df['거래량'].std() #표준편차
df['거래대금'].var() #분산
df['저가'].median() #중앙값
df['시가'].skew() #왜도
df['고가'].kurt() #첨도
df.mean(numeric_only=True) #숫치열의 평균만 구함함
df['계산결과'] = df['종가']*df['거래량']
df['회사이름'] = df['종목명'] + '회사' #시리즈 추가
print(df['회사이름'])

display(df)
```

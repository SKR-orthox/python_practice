```python
import numpy as np
import pandas as pd
import seaborn as sns #seaborn 모듈 호출

fmri = sns.load_dataset('fmri') #fmri 데이터셋 가져오기
sns.set_style('whitegrid') #하얀 바탕에 그리드 생성
sns.lineplot(x='timepoint', y='signal',data=fmri) #선그래프 작성

fmri.sample(n=10, random_state=1) #fmri에서 랜덤하게 10개의 값 추출(값 보고싶을 땐 위아래 주석처리리)

sns.lineplot(x='timepoint', y= 'signal', hue='event', data=fmri)
```

```python
#회귀 그래프, 회귀선을 포함한 산점도
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')

#'data='에는 사용할 데이터셋 지정, 'x_ci=95'는 x축에 대한 95% 신뢰 구간(confidence interval, 하늘색 영역역)
sns.regplot(x='total_bill', y='tip', data=tips, ci=95) 
```

```python
#산점도
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
sns.scatterplot(x='total_bill', y= 'tip', hue="time", data=tips)
```

```python
#비교 그래프
import seaborn as sns

tips = sns.load_dataset('tips')
sns.countplot(x='smoker', hue='time', data= tips)
```

```python
#막대그래프
#일반적인 막대 그래프는 데이터의 개수를 표시하는 데 반해, 시본의 막대그래프는 y 값이 연속형 값일 경우 해당 값들의 평균값을 나타냄, 신뢰구간은 검은색 막대로 표현
import seaborn as sns

tips = sns.load_dataset('tips')
sns.barplot(x='day', y='total_bill', data=tips)
#tips.sample(n=10, random_state=1) #파일 내용 확인
```

```python
#분포도 그리기
import seaborn as sns

tips = sns.load_dataset('tips')
#sns.scatterplot(x='total_bill', y='tip', data=tips)
sns.relplot(x='total_bill', y='tip', data=tips) #scatterplot으로도 그릴 수 있으나, 관계 비교할 때 replot이 특화
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker') #흡연자로 비교교
sns.relplot(x='total_bill', y='tip', data=tips, hue='size') #사이즈별 비교
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='Blues')#색상 변경
```

```python
#선형 그래프
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))

fmri = sns.load_dataset('fmri')

sns.relplot(x='time', y='value', data=df, kind='line')
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci=None)
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci='sd') #신뢰구간 표준편차
```

```python
#범주형 그래프
import seaborn as sns


tips = sns.load_dataset('tips')
#display(tips)
#sns.catplot(x='day', y='total_bill', data=tips)
#sns.catplot(x='day', y='total_bill', data=tips, kind='swarm') #마커가 겹치지 않도록 swarm
#sns.catplot(x='day', y='total_bill', data=tips, kind='box') 
#sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')#box 에서 더 나눈 것
sns.catplot(x='day', y='total_bill', data=tips, kind='violin') #마커가 겹치지 않도록 swarm
#sns.catplot(x='day', y='total_bill', data=tips, kind='bar')#선은 표준편차
#sns.catplot(x='day', y='total_bill', data=tips, kind='point')

#inner='stick':바이올린 내부에 각 관측치마다 짧은 선(스틱)을 그려서 개별 데이터 분포 표시
#split=True: hue가 딱 2개 범주일 때, 하나의 바이올린을 좌우 반쪽으로 나눠 두 범주를 한 자리에서 비교, hue가 없거나 범주가 2개 초과면 동작하지 않거나 무시
sns.catplot(x='day', y='total_bill', data=tips, inner='stick', split =True, kind='violin', hue='sex')
```

```python
#그래프 합치기
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.randn(500,2), columns=['1','2'])
display(df)
sns.relplot(x='1', y='2', data=df) #분포도도
sns.jointplot(x='1', y='2', data=df) #정규화된 데이터 확인
sns.jointplot(x='1', y='2', data=df, kind='hex') #육각형
sns.jointplot(x='1', y='2', data=df, kind='kde') #등고선
sns.jointplot(x='1', y='2', data=df, kind='reg') #평균값을 사용한 회귀 그래프
sns.jointplot(x='1', y='2', data=df, kind='resid') #reg와 유사하지만 편자 표현 x
```

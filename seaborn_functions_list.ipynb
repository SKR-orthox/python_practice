```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #seaborn 모듈 호출

fmri = sns.load_dataset('fmri') #fmri 데이터셋 가져오기
sns.set_style('whitegrid') #하얀 바탕에 그리드 생성
sns.lineplot(x='timepoint', y='signal',data=fmri) #선그래프 작성

fmri.sample(n=10, random_state=1) #fmri에서 랜덤하게 10개의 값 추출

sns.lineplot(x='timepoint', y= 'signal', hue='event', data=fmri)
```

```python
#회귀 그래프, 회귀선을 포함한 산점도
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=tips, ci=95) #'data='에는 사용할 데이터셋 지정, 'x_ci=95'는 x축에 대한 95% 신뢰 구간
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #seaborn 모듈 호출

fmri = sns.load_dataset('fmri') #fmri 데이터셋 사용
sns.set_style('whitegrid') #하얀 바탕에 그리드 생성
sns.lineplot(x='timepoint', y='signal',data=fmri) #선그래프 작성
```
